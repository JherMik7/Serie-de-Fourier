```python
#Importamos las librerias
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp
from matplotlib import style
plt.style.use(['science', 'notebook'])
```

```python
# Definimos una funcion que calcule la suma total de los terminos
def fourier(x):
    suma = 0 # Inicializamos el valor de la suma
    for n in range(1, 16000, 1): # Bucle para los valores de n 
        suma += (2/np.pi)*(1-(-1)**n)*np.sin(n*x)/n # Formula de la serie de Fourier
    return suma # Valor que retorna la funcion Fourier

# Definimos un dominio:
x = np.arange(-np.pi, np.pi, 0.001)
F = fourier(x) # Guardamos en un vector F los valores de evaluados del dominio

plt.plot(x, F, color = 'springgreen') # Graficamos
#plt.savefig('fourier.eps') # Guardamos la grafica
```

```python
l = 3
# Definimos una funcion que calcule la suma total de los terminos
def fourier(x):
    suma = 0 # Inicializamos el valor de la suma
    for n in range(1, 90, 1): # Bucle para los valores de n 
        suma += (2*l/np.pi)*((-1)**(n+1))*np.sin(np.pi*n*x/l)/n # Formula de la serie de Fourier
    return suma # Valor que retorna la funcion Fourier

dx = 0.005
# Definimos un dominio:
x = np.arange(-l, l+dx, dx)
F = fourier(x) # Guardamos en un vector F los valores de evaluados del dominio

plt.plot(x, F, color = 'violet') # Graficamos
#plt.grid(True) # Activamos la grilla
#plt.show() # Mostramos el grafico
```
